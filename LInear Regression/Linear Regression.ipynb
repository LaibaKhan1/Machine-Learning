# Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hello.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.Area,df.Price,color = "red", marker = "+")
```

```python
#new data frame
#imsein price k column ko drop krdengy 
#liner model ko jb hum object bana kr call krty hain tw wo expect krta hai 2D array  
#yea naya df (data frame) hai yea bhi 2D array hai sirf area ka column 

new_df = df.drop("Price", axis= "columns")
new_df
```

```python
price = df.Price
price
```

```python
reg = linear_model.LinearRegression()
#model #yea ek es class ka object create hgya hai 

#is model ko train krny k liye 
reg.fit(new_df,price)
```

```python
#Predict price of a home with area = 3300 sqr ft

reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
